```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
mona = "images/Mona Lisa.jpg"
mona2 = "images/Mona Lisa2.jpg"
creation = "images/Creation of Adam.jpg"
```

```python
# def cross_stochify(path1, path2):
#     # read in images
#     img1 = Image.open(path1)
#     img2 = Image.open(path2)
    
#     # convert to grayscale
#     gray1 = np.array(img1.convert("L"))
#     gray2 = np.array(img2.convert("L"))
    
#     # Check if "raw" matrices are able to be multiplied
#     if gray1.shape[1] == gray2.shape[0]:
#         stoch = np.dot(gray1, gray2)
#     elif gray1.shape[0] == gray2.shape[1]:
#         stoch = np.dot(gray2, gray1)
#     else:
#         # if "raw" matrices CANNOT be multiplied, then trim
#         if gray1.shape[1] > gray2.shape[0]:
#             gray1 = gray1[:, :gray2.shape[0]]  # trim columns of A to match the number of rows in B
#             stoch = np.dot(gray1, gray2)
#         else:
#             gray2 = gray2[:gray1.shape[0], :]  # trim rows of B to match the number of columns in A
#             stoch = np.dot(gray1, gray2)
    
#     # display the resulting image
#     name1 = path1.split('.')[0]
#     name2 = path2.split('.')[0]
#     plt.imshow(stoch, cmap='gray')
#     plt.title(f"{name1} \u00A7 {name2}")
#     plt.show()

# # Example usage:
# # stochify("path/to/image.jpg")
# # self_stochify("path/to/image.jpg")
# # cross_stochify("path/to/image1.jpg", "path/to/image2.jpg")

def cross_stochify(path1, path2):
    # read in images
    img1 = cv2.imread(path1, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(path2, cv2.IMREAD_GRAYSCALE)
    
    # Check if "raw" matrices are able to be multiplied
    if img1.shape[1] == img2.shape[0]:
        stoch = np.dot(img1, img2)
    elif img1.shape[0] == img2.shape[1]:
        stoch = np.dot(img2, img1)
    else:
        # if "raw" matrices CANNOT be multiplied, then trim
        if img1.shape[1] > img2.shape[0]:
            img1 = img1[:, :img2.shape[0]]  # trim columns of A to match the number of rows in B
            stoch = np.dot(img1, img2)
        else:
            img2 = img2[:img1.shape[0], :]  # trim rows of B to match the number of columns in A
            stoch = np.dot(img1, img2)
    
    # display the resulting image
    name1 = path1.split('.')[0]
    name2 = path2.split('.')[0]
    plt.imshow(stoch, cmap='gray')
    plt.title(f"{name1} \u00A7 {name2}")
    plt.show()

cross_stochify(mona,creation)
```
